```python
import pandas as pd
import numpy as np
import matplotlib as plt
from matplotlib import pyplot
df = pd.read_csv('jsrt_metadata.csv')
df
df['diagnosis'].value_counts()
plt.pyplot.title('Diagnosis')
plt.pyplot.xlabel('Number of Disease')
plt.pyplot.ylabel('Patients')
plt.pyplot.hist(df['diagnosis'].value_counts())
```

Question # 2

```python
for a in list(df['diagnosis'].unique()):
    plt.pyplot.figure()
    plt.pyplot.pie(df.loc[df['diagnosis'] == a]['state'].value_counts(), labels = list(df.loc[df['diagnosis'] == a]['state'].value_counts().keys()), autopct = '%1.1f%%',colors=['yellow','lightblue'])
    plt.pyplot.legend(title= a)
    print(df.loc[df['diagnosis'] == a]['state'].value_counts())
```
